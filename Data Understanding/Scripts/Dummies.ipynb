```python
import pandas as pd

def load_original_data(filename):
    """
    Load data from a CSV file into a DataFrame.

    :param filename: The name of the CSV file to load.
    :return: A pandas DataFrame containing the data.
    """
    original_df = pd.read_csv(filename)
    return original_df

def create_dummy_dataframes(original_df, num_dummies):
    """
    Create a list of dummy DataFrames with the same structure as the original DataFrame.

    :param original_df: The original DataFrame to mimic.
    :param num_dummies: The number of dummy DataFrames to create.
    :return: A list of dummy DataFrames.
    """
    dummy_dfs = [original_df.copy() for _ in range(num_dummies)]
    return dummy_dfs

def modify_dummy_dataframes(dummy_dfs):
    """
    Modify the values in the dummy DataFrames by applying an arithmetic operation.

    :param dummy_dfs: A list of dummy DataFrames to modify.
    """
    for dummy_df in dummy_dfs:
        for column in dummy_df.columns:
            # Apply an arithmetic operation to modify values, excluding the first row
            if column != "time":
                dummy_df[column] = dummy_df[column].apply(lambda x: x * 2)

def save_dummy_dataframes(dummy_dfs):
    """
    Save the modified dummy DataFrames as separate CSV files.

    :param dummy_dfs: A list of modified dummy DataFrames.
    """
    for i, dummy_df in enumerate(dummy_dfs):
        dummy_df.to_csv(f'dummy_{i + 1}_modified.csv', index=False)

# Main part of the script
if __name__ == "__main__":
    # Load the original data
    original_df = load_original_data('original_data.csv')

    # Create dummy dataframes
    num_dummies = 3
    dummy_dfs = create_dummy_dataframes(original_df, num_dummies)

    # Modify the dummy dataframes
    modify_dummy_dataframes(dummy_dfs)

    # Save the modified dummy dataframes
    save_dummy_dataframes(dummy_dfs)
```
